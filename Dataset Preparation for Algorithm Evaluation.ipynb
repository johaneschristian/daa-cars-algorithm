```python
import numpy as np
import math
import pandas as pd
from sklearn import model_selection as ms
```

```python
ratings = pd.read_csv('ratings.csv')

# Untempered dataframe
original_ratings = ratings.dropna()

# Only consider data with complete context
ratings = ratings.dropna()

ratings.info
```

```python
users = ratings['userid'].unique()
```

```python
kfold = ms.KFold(10, shuffle=True, random_state=1)
kfold.split(users)

witheld_data_collection = []

for train, test in kfold.split(users):
    test_set = users[test]
    break

for user in test_set:
    user_rating = ratings[ratings['userid'] == user]
    row_count = user_rating.shape[0]
    
    twenty_percent = math.ceil(row_count * 0.2)
    print('Number of items dropped', twenty_percent)

    # Get 20% of the tested group
    witheld_data = user_rating.tail(twenty_percent)
    witheld_data_collection.append(witheld_data)
    
    display(witheld_data)
    
    ratings.drop(
        witheld_data.index,
        inplace=True
    )

display(ratings.info)

witheld_data_collection[0]
```
