```python
import pandas as pd
import numpy as np
```

```python
ratings = pd.read_csv('ratings_noNAN.csv')
display(ratings)
```

```python
ratings_1082 = ratings[ratings['userid'] == '1082']
```

```python
display(ratings_1082)
```

```python
ratings_1082 = ratings_1082.sample(frac=1)
```

```python
display(ratings_1082)
```

```python
witheld_ratings = ratings_1082.tail(153)
```

```python
witheld_ratings.sort_values(
    by='rating', 
    ascending=False,
    inplace=True)

display(witheld_ratings)
```

```python
witheld_ratings.groupby('rating').count()
```

```python
witheld_ratings.index
```

```python
training_dataset = ratings.drop(witheld_ratings.index)
```

```python
training_dataset.to_csv('training_dataset.csv', index=False)
```
